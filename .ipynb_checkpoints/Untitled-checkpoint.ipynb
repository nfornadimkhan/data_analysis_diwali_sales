# Exploratory Data Analysis

## Gender

```python
df.columns
```

```python
# Create the countplot
ax = sns.countplot(x='Gender', data=df, palette='Set2', hue='Gender')

# Add labels to each bar
for container in ax.containers:
    ax.bar_label(container)

# Show the plot
plt.show()
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

From the above graph it is clear that most of the buyers are females.

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x= 'Gender', y= 'Amount', data=sales_gen, palette='Set2', hue='Gender')
plt.show()
```

From the above graph it is clear that the purchasing power of females are greater than men.

## Age

```python
ax = sns.countplot(data=df, x= 'Age Group', hue='Gender')

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Age Group', y='Amount',data=sales_age, hue='Age Group')

plt.show()
```

From the above two graphs it clear that most of buyers are of age group between 26 to 35 years females

## State

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x = 'State', y = 'Orders', hue='State')
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(8)
sns.set(rc={'figure.figsize': (15,5)})
sns.barplot(data=sales_state, x= 'State', y= 'Amount', hue='State')
plt.show()
```

From the above graph we can conclude that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively.

## Marital Status

```python
ax = sns.countplot(data=df, x='Marital_Status', hue='Marital_Status')

sns.set(rc={'figure.figsize':(4,5)})
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount', hue='Gender')
plt.show()
```

From above graphs we can see that most of the buyers are unmarried (women) and they have hight purchasing power.

## Occupation

```python
ax = sns.countplot(data=df, x='Occupation', hue='Occupation')
sns.set(rc={'figure.figsize':(20,5)})

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_occupation, x='Occupation', y='Amount', hue='Occupation')

plt.show()
```

From the above graph it can be conlcluded that most of the buyers are working in IT, Healthcare and Aviation sector.

## Product Category

```python
ax = sns.countplot(data=df, x= 'Product_Category', hue='Product_Category')
sns.set(rc={'figure.figsize':(20,5)})

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
sales_product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_product, x = 'Product_Category', y='Amount', hue='Product_Category')
plt.show()
```

From the above graph we can conclude that most of the sold products are from Food, Clothing & Apparel and Electronics & Gadgets.

```python
# 10 most sold products
sales_product_ID = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_product_ID, x = 'Product_ID', y='Orders', hue='Product_ID')
plt.show()
```

## Conclusion

- **Demographics of Buyers:** The majority of buyers are females, particularly those aged 26 to 35 years, indicating that this demographic has the highest purchasing power.
- **Geographical Insights:** The highest number of orders and total sales originate from Uttar Pradesh, Maharashtra, and Karnataka, emphasizing these as key markets.
- **Marital Status and Purchasing Power:** Unmarried women represent a significant segment of buyers with high purchasing power.
- **Occupational Trends:** Most buyers work in the IT, Healthcare, and Aviation sectors, highlighting these industries as lucrative customer bases.
- **Product Preferences:** The most purchased products fall into the categories of Food, Clothing & Apparel, and Electronics & Gadgets, suggesting these categories should be prioritized for inventory and marketing efforts.
